## Exercise #1 <br>
<p>Cube Number Test... Print out all cubed numbers up to the total value 1000. Meaning that if the cubed number is over 1000 break the loop.</p>

```python
for num in range(0,1001):
    if num ** 2 > 1000:
        break
    else:
        print(num ** 2)
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
# HINT::
# An else after an if runs if the if didnt
# An else after a for runs if the for didnt break
numbers = range(2,101)
not_primes = []
primes = []

for num1 in numbers:
    for num2 in numbers:
        if num1 * num2 not in not_primes and num1 * num2 < 100:
            not_primes.append(num1 * num2)
for i in range (2,101):
    if i not in not_primes:
        primes.append(i)
print(primes)

        
            
    
    

```

# Exercise 3 <br>
<p>Take in a users input for their age, if they are younger than 18 print kids, if they're 18 to 65 print adults, else print seniors</p>

```python
age = int(input("How old are you?"))

if age <= 0:
    print("Please put a valid age")
elif age < 18:
    print("kids")
elif age >=18 and age <=65:
    print("adults")
elif age > 65 and age < 9000:
    print("seniors")
elif age == 9000:
    print("Shoha")
else: 
    print(f"Shoha in {age - 9000} years")
```
